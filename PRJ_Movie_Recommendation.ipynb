## Demographic Filtering
Demographic Recommender system generate recommendations based on the user demographic attributes. It categorize the users based on their attributes and recommends the movies by utilizing their demographic data.
They offer generalized recommendations to every user, based on movie popularity and/or genre. The System recommends the same movies to users with similar demographic features. Since each user is different , this approach is considered to be too simple. 
We can use the average ratings of the movie as the score but using this won't be fair enough since a movie with 8.9 average rating and only 3 votes cannot be considered better than the movie with 7.8 as as average rating but 40 votes. So, I'll be using IMDB's weighted rating (wr) which is given as :-

where,

v is the number of votes for the movie;

m is the minimum votes required to be listed in the chart;

R is the average rating of the movie;

C is the mean vote across the whole report

```python
C = movies_data['vote_average'].mean()
C
```

So, the mean rating for all the movies is approx 6 on a scale of 10.Now we will determine an appropriate value for m, the minimum votes required to be listed in the chart. We will use 90th percentile as our cutoff so that we can get top 10 best movies from the list.

```python
m= movies_data['vote_count'].quantile(0.9)
m
```

```python
q_movies = movies_data.copy().loc[movies_data['vote_count'] >= m]
q_movies.shape
```

Here we can see that the dataset has been filtered out based on the criteria we mentioned giving us 481 movies qualified

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies['score']
```

```python
#Sort movies based on score calculated above
q_movies = q_movies.sort_values('score', ascending=False)

#Print the top 15 movies
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(15)
```

## Collaborative Filtering
In Collaborative Filtering, we tend to find similar users and recommend what similar users like. In this type of recommendation system, we don’t use the features of the item to recommend it, rather we classify the users into the clusters of similar types, and recommend each user according to the preference of its cluster.
To address some of the limitations of content-based filtering, collaborative filtering uses similarities between users and items simultaneously to provide recommendations.



##### Item Based Filtering
Item-based collaborative filtering is a recommendation system that recommends items to users based on the similarity between items. The system creates an item-item similarity matrix and recommends items that are similar to the ones the user has already liked or rated highly.

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

#### Loading the Movie Dataset we have chosen

```python
# Load the movies dataset
movies_df = pd.read_csv('movies.csv')

# Drop any rows with missing values
movies_df.dropna(inplace=True)
```

The TF-IDF (Term Frequency-Inverse Document Frequency) is an algorithm that converts a collection of raw documents into a matrix of TF-IDF features. The matrix represents each document as a vector of features, where each feature is a word from the text, and the value of the feature represents the importance of the word in the document and the entire corpus.

 It takes in a corpus of text documents, applies text preprocessing techniques, such as tokenization and stemming, and then calculates the TF-IDF weights for each word in each document. These weights are then used to represent the text documents as vectors of features.

```python
# Create a TF-IDF vectorizer object
tfidf = TfidfVectorizer(stop_words='english')

# Computing the TF-IDF matrix for the genres
tfidf_matrix = tfidf.fit_transform(movies_df['genres'])
```

Cosine similarity is a metric used to measure the similarity between two vectors in a high-dimensional space.

It is a useful metric for comparing the similarity between text documents, where each document is represented as a vector of TF-IDF features. It can also be used to cluster similar documents, recommend similar products, and detect anomalies in large datasets.






The similarity score is a measure of the similarity between the two vectors being compared, based on their magnitudes and the cosine of the angle between them. The closer the similarity score is to 1, the more similar the vectors are, while a score of 0 or -1 indicates that the vectors are dissimilar or opposite.

```python
# Computing the cosine similarity matrix for the TF-IDF matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)


def get_similar_movies(movie_title, N=10):
    
    # Get the index of the movie with the given title
    idx = movies_df[movies_df['title'] == movie_title].index[0]

    # Get the cosine similarity scores for all movies compared to the given movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies by their similarity score, in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Getting the top N most similar movies
    sim_scores = sim_scores[1:N+1]
    movie_indices = [i[0] for i in sim_scores]
    return movies_df['title'].iloc[movie_indices]
```

```python

movie_title = input("Enter a movie title: ")

# Check if the movie title exists in the dataframe

if movie_title not in movies_df['title'].values:
    print("Sorry, we could not find a movie with that title.")
else:
    
    similar_movies = get_similar_movies(movie_title, N=10)

    # Giving the users a recommended list of movies
    print(f"Top 10 movies similar to '{movie_title}':")
    print(similar_movies)
```

## Knowledge Based Filtering

 Knowledge-based filtering, also known as content-based filtering, is a type of recommendation system that suggests items to users based on their past behavior and preferences. It works by analyzing the features of items and comparing them to the user's past preferences and behavior to make recommendations. The system uses information about the items, such as genre, director,and actors, to identify items that are similar to those the user has enjoyed in the past.
 
 The system does not rely on the behavior or preferences of other users, and can provide recommendations even for new users who have not yet provided any feedback.This is the major advantage of Knowledge based Filtering

```python
import pandas as pd

# Load the movies dataset
movies_df = pd.read_csv('movies.csv')
```

```python
# Drop any rows with missing values
movies_df.dropna(inplace=True)
```

Asking Users the criteria for the basis of their desired recommendation.

```python
def recommend_movies():
    
    # Ask user for their preferred genre
    genre = input("What genre of movie are you interested in? ")
    
    # Filter the movies dataset by genre
    genre_movies = movies_df[movies_df['genres'].str.contains(genre, case=False)]
    
    # Ask user for their preferred actor/actress
    actor = input("Which actor or actress would you like to see in the movie? ")
    
    # Filter the movies dataset by actor/actress
    actor_movies = genre_movies[genre_movies['cast'].str.contains(actor, case=False)]
    
    # Ask user for their preferred director
    director = input("Who should direct the movie? ")
    
    # Filter the movies dataset by director
    director_movies = actor_movies[actor_movies['director'].str.contains(director, case=False)]
    
    
    
    if len(director_movies) > 0:
        top_movies = director_movies.sort_values(by='popularity', ascending=False).head(10)
        print("Here are the top 10 movies that match your preferences:")
        for index, row in top_movies.iterrows():
            print(row['title'])
    else:
        print("Sorry, we could not find any movies that matches your preferences.")


recommend_movies()
    
```
